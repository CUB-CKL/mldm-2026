```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(123456789)
```

```python
n1 = 128
n2 = 128 + 64

X1 = rng.uniform(size=(n1, ))
X2 = rng.uniform(size=(n2, ))

eps1 = rng.normal(size=(n1, ))
eps2 = rng.normal(size=(n2, ))

m1 = np.sqrt(X1)
m2 = np.square(X2)

sigma_1 = np.exp(-2 - 2 * X1)
sigma_2 = np.exp(2 * X2 - 4)

y1 = m1 + eps1 * sigma_1
y2 = m2 + eps2 * sigma_2

X = np.concatenate([X1, X2], axis=0)
y = np.concatenate([y1, y2], axis=0)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.savefig('../01-lecture/imgs/tricky-data.png')
```

```python
xs = np.linspace(0, 1, num=128)
m1 = np.sqrt(xs)
s1 = np.exp(-2 - 2 * xs)
m2 = np.square(xs)
s2 = np.exp(2 * xs - 4)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(xs, m1, color=plt.cm.tab10(0))
plt.plot(xs, m2, color=plt.cm.tab10(1))

plt.fill_between(xs, m1 - s1, m1 + s1, color=plt.cm.tab10(0), alpha=0.5)
plt.fill_between(xs, m2 - s2, m2 + s2, color=plt.cm.tab10(1), alpha=0.5)

plt.xlabel('X')
plt.ylabel('y')
plt.savefig('../01-lecture/imgs/tricky-trends.png')
```

```python
Xy = np.concatenate([X, y], axis=1)
np.savetxt('data.txt', Xy, )
```
